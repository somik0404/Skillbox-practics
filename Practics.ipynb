```python
# Импортируем нужные модули
import pandas as pd
import numpy as np
import copy
import matplotlib.pyplot as plt 
import datetime as dt
import seaborn as sns
```

```python
# Загружаем нужные файлы
progress_ph = pd.read_csv("progress_phases.csv")
progress = pd.read_csv("progresses.csv")
course_con = pd.read_csv("course_contents.csv")
courses = pd.read_csv("courses.csv")

# Объединяем в один большой датафрейм main
df = progress_ph.merge(progress, left_on="progress_id", right_on="id")
df = df.drop(["id"], axis="columns")
df = df.merge(courses, left_on="course_id", right_on="id")
df = df.drop(["id", "Unnamed: 0"], axis="columns")
df = df.merge(course_con, on=["module_number", "lesson_number", "course_id"])
df.head()
```

### Минимальный, максимальный, средний, медианный возраст студентов

```python
# Загружаем Датафрейм с датами рождения студентов
students = pd.read_csv("students.csv")

# Фильтруем его
students.dropna().set_index("id_", inplace=True)
students = students.drop("id_", axis="columns")
students['birthday'] = pd.to_datetime(students["birthday"], errors='coerce')
students["age"] = datetime.now()
students["age"] = (students["age"] - students["birthday"]).dt.days // 365
students = students[students["age"] < 100]
students = students[students["age"] > 0]

# Выводим все нужные значения
print("Минимальный вораст -- ", students['age'].min())
print("Максимальный вораст -- ", students['age'].max())
print("Средний вораст -- ", students['age'].mean())
print("Медианный вораст -- ", students['age'].median())
```

### Минимальный, максимальный, средний, медианный возраст студентов на каждом курсе

```python
# Объединяем Датафрейм с датами рождения с основным
students_ages = df.merge(students, left_on="student_id", right_on="id")
students_ages = students_ages.drop("id", axis="columns")

# Грпуппируем данные по названию курса и студентов, которые на нём учатся
students_ages = students_ages.groupby(["title", "student_id"]).agg("first")

# Высчитыеваем необходимые значения агрегацией
students_ages = students_ages.groupby(level=0).agg({"age":[np.min, np.max, np.median]})
students_ages.head()
```

### Количество студентов на каждом курсе (bar-chart)

```python
# Считаем количество студентов на каждом курсе
total_students = df.groupby(['title']).nunique()
total_students = total_students.sort_values("student_id", ascending=False)

# Рисуем диаграму по этим значениям
plt.figure(figsize=(14, 7))

# Украшаем таблицу
plt.title("Количесво студентов на каждом курсе", fontsize=16)
sns.barplot(x=total_students.index, y=total_students["student_id"])
plt.xticks(rotation=90)
plt.xlabel("Название курса", fontsize=16)
plt.ylabel("Количество студентов", fontsize=16)
```

### Постройте горизонтальный bar-chart, отражающий количество студентов на каждом курсе. График должен иметь заголовок. Значения должны быть отсортированы. Нужна легенда. Прозрачность - 0.1. На график должна быть нанесена линия медианы, со своим цветом. Рамки не должно быть

```python
# Считаем количество студентов на каждом курсе
test = df.groupby("title").agg({"student_id": "nunique", "field": "first"}).sort_values(by="student_id", ascending=True)

# Создадим словарь цветов для каждой сферы
mapper = sns.color_palette("deep", test.field.unique().size)
color_dict = {}
for num, field in enumerate(test.field.unique()):
    color_dict[field] = mapper[num]

# Формируем палитру для графика
my_pallete=[]
for element in test.field:
    my_pallete.append(color_dict[element])

# Рисуем само поля для графика и его самого
plt.figure(figsize=(20,10))
students_hbar = sns.barplot(x="student_id", y=test.index, data=test, palette=my_pallete, alpha=0.9)

# Рисуем вертикальную линию по медианному количеству студенто и добавляем для неё отметку и легненду
fl = plt.axvline(test.student_id.median(), ls="--", lw=3, color="purple", alpha=0.9, label="Медианное количество студентов")
xt = students_hbar.get_xticks()
xt = np.append(xt, test.student_id.median())
students_hbar.set_xticks(xt)
first_legend = plt.legend(handles=[fl], loc=7, fontsize=20, shadow=True)

# Добавим эту легенду в текущий axes
plt.gca().add_artist(first_legend)

# Нарисуем прямоугольники для легенды и присвоим им цвета из словавря
handles = [plt.Rectangle((0,0), 1, 1, color=color_dict[key], alpha=0.9) for key in color_dict.keys()]

# Отрисуем вторую легенду
plt.legend(handles, color_dict.keys(), fontsize=20, title="Field", title_fontsize=20)

# Рисуем подписи для графика и осей
plt.title("Количество студентов по курсам", fontsize=25)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.xlabel("Количество студентов", fontsize=20)
plt.ylabel(None)

# Убираем рамку и добавляем сетку и отрисовываем получившуюся диаграмму
plt.box(False)
plt.grid(axis="x", alpha=0.3, ls="--", lw=2)
plt.show()
```

### Рассчитать прирост студентов на каждом курсе в каждом месяце за всю историю. Считать дату начала прохождения курса студентом по дате начала первой домашней работы.

```python
# Создаём копию таблицы
increase_students = copy.deepcopy(df)

# Выбираем уроки, только где есть ДЗ
increase_students = increase_students[increase_students['is_homework'] == True]

# Ищем первые уроки в каждом модуле
increase_students = increase_students.groupby(['title', 'module_number']).agg({'lesson_number':'min'})

# Выводим таблицу с модулем, в котором есть первый урок с ДЗ
increase_students = increase_students.groupby(['title', 'module_number']).first()
increase_students = increase_students.reset_index()
increase_students = increase_students.groupby('title').first()

# Соединяем эту таблицу с общей
increase_students = increase_students.merge(df, on=['title', 'module_number', 'lesson_number'], how='left')

# Оставляем только месяц и год
increase_students['start_date'] = pd.to_datetime(increase_students['start_date'].str[0:7])

# Группируем и смотрим, когда ученик начал выполнять первое ДЗ
increase_students = increase_students.groupby(['title', 'student_id']).agg({'start_date':'min'})

# Снова оставляем месяц и год в дате
increase_students = increase_students.reset_index()
increase_students = increase_students.astype({'start_date': 'str'})
increase_students['start_date'] = increase_students['start_date'].str[0:7]

# Считаем количество студентов в каждом месяце
increase_students = increase_students.groupby(['title', 'start_date']).agg({'student_id':'count'})
increase_students = increase_students.reset_index()

# Строим сводную таблицу
increase_students = increase_students.pivot(index='start_date', columns='title', values='student_id').fillna(0)

increase_students
```

### На основании первого пункта построить line-graph с приростом студентов в каждом месяце для каждого курса. 15 графиков. Графики должны иметь заголовки, оси должны быть подписаны. Ticks нужно развернуть так, чтобы они были читаемы.

```python
# Создаём копию таблицы и меняем там тип
graph = copy.deepcopy(increase_students)
graph.reset_index(inplace=True)

# Создаём список всех столбцов
list_courses = list(graph.columns)[1:]

# Строим графики
for course in list_courses:
    fig, ax = plt.subplots(figsize=(12,3))
    ax.plot(graph["start_date"], graph[course])
    ax.set_title(f'Прирост студентов на курсе {course}')
    ax.set_ylabel("Прирст студентов")
    plt.xticks(rotation=45)
    plt.grid()
```

### Рассчитать количество прогрессов по выполнению домашних работ в каждом месяце за всю историю для каждого курса. Учитывать, что выполнение домашнего задания может перетекать из одного месяца в другой (такие дз надо включать в общее число прогрессов для всех месяцев, которые покрывает срок выполнения этих дз)

```python
# Создаём копию датафрейма
progress = copy.deepcopy(df)

# Подготавливаем датафрейм для дальнейшей работы с ним
progress = progress[progress['status'] == 'done']
progress = progress[progress['is_homework'] == True]
progress['start_period'] = pd.to_datetime(progress['start_date'].str[0:7])
progress['end_period'] = pd.to_datetime(progress['finish_date'].str[0:7])

# Создаём новый датафрейм
count_progress = {'title':[],'progress_id':[], 'period':[]}

# Пишем формулу, по которой будем считать прогресс
def count_progress_all(title, start_period, end_period, progress_id):
    if(start_period.year == end_period.year & start_period.month == end_period.month):
        list_dt.append(start_period)
    else:
        list_dt = pd.period_range(start=start_period, end=end_period, freq='M') 
    for dt in list_dt:
        count_progress['title'].append(title)
        count_progress['progress_id'].append(progress_id)
        count_progress['period'].append(dt)

# Применяем эту формулу
progress.apply(lambda row: count_progress_all(row.title, row.start_period, row.end_period, row.progress_id), axis=1)

# Работаем с новым датафреймом
df_count_progress=pd.DataFrame(count_progress)
df_count_progress = df_count_progress.groupby(['title', 'period']).agg({'progress_id':'count'})
df_count_progress.reset_index(inplace=True)

# Делаем сводную таблицу по нему
df_count_progress = df_count_progress.pivot(index='period', columns='title', values='progress_id').fillna(0)
df_count_progress.index = df_count_progress.index.astype('str')

# Выводим таблицу, чтобы посмотреть на результат
df_count_progress.head()
```

### Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле, к количеству студентов, сдавших задание в предыдущем модуле

```python
# Считаем количество начавших выполнять задания на каждом модуле курса
start_df = copy.deepcopy(df)
start_df = start_df[start_df['is_homework'] == True]
start_df = start_df.groupby(['title', 'module_title', 'module_number'])['start_date'].count().reset_index()\
                 .sort_values(['title', 'module_number'])
start_df = start_df.reset_index()
start_df = start_df.drop('index', axis='columns')

# Считаем количнество закончивших выполнять задание на каждом модуле курса 
finish_df = copy.deepcopy(df)
finish_df = finish_df[finish_df['is_homework'] == True]
finish_df = finish_df.groupby(['title', 'module_title', 'module_number'])['finish_date'].count().reset_index()\
                 .sort_values(['title', 'module_number'])
finish_df = finish_df.groupby(['title', 'module_title', 'module_number']).agg({"finish_date":"first"}).sort_values(['title', 'module_number'])

# Шифтим строчки на один вниз
finish_df = finish_df.shift(periods=1, fill_value=0)
finish_df = finish_df.reset_index(drop=True)

# Объединяем датафреймы
total_df = pd.concat([start_df, finish_df], axis=1)

# Находим первые модули в каждом модуле
first_line = total_df.groupby('title').first()
first_line = first_line.reset_index()

# Объединаем добщий датафрейм и датафреймом из первых строчек и убираем дубликаты
data = pd.concat([total_df, first_line]).drop_duplicates(keep=False)
data = data.sort_values(['title', 'module_number'])

# Делаем столбец с конверсией
data['convertion'] = data['start_date'] / data['finish_date']

# Смотрим на результат
data
```
